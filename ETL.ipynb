### 1-EN PRIMERA INSTANCIA, DESARROLLAMOS EL ETL DE LAS TABLAS 

```python
############################################ VENTAS_POR_DIAS #######################################################
```

```python
#Levantamos los datasets
Sales_for_date = pd.read_csv("VENTAS_POR_DIAS.CSV",sep=',')
Sales_for_date
```

```python
#En la columna 'fecha', eliminamos las primeras tres letras
Sales_for_date['fecha'] = Sales_for_date['fecha'].str.slice(3)
Sales_for_date
```

```python
# Eliminamos las columnas 'cubt1', 'cubt2', 'cub' y 'promedio'
Sales_for_date = Sales_for_date.drop(['cubt1', 'cubt2', 'cubt2','cub','promedio'], axis=1)
Sales_for_date
```

```python
#Eliminamos las columnas 'tickets', 'cubt3'.
Sales_for_date = Sales_for_date.drop(['tickets', 'cubt3'], axis = 1)
Sales_for_date
```

```python
Sales_for_date.info()
```

```python
#Cambio tipo de dato " efectivo", "cta_cte","tarjetas", "cheques", "otros", "total", "prom_vtas", "acumulado"
Sales_for_date['efectivo'] = Sales_for_date["efectivo"].astype("float")
Sales_for_date['cta_cte'] = Sales_for_date["cta_cte"].astype("float")
Sales_for_date['tarjetas'] = Sales_for_date["tarjetas"].astype("float")
Sales_for_date['cheques'] = Sales_for_date["cheques"].astype("float")
Sales_for_date['otros'] = Sales_for_date["otros"].astype("float")
Sales_for_date['total'] = Sales_for_date["total"].astype("float")
Sales_for_date['prom_vtas'] = Sales_for_date["prom_vtas"].astype("float")
Sales_for_date['acumulado'] = Sales_for_date["acumulado"].astype("float")
Sales_for_date.info()
```

```python
#Observamos los primeros 5 datos 
Sales_for_date.head()
```

```python
Sales_for_date.to_csv('Sales_for_date.csv', index=False)
```

```python
################################################### VENTAS_POR_CLIENTE #############################################
```

```python
sales_by_customers = pd.read_csv ("VENTAS_POR_CLIENTE.CSV", sep="," , encoding='latin-1')
sales_by_customers
```

```python
# Verificamos las columnas con cantidad de valores nulos 
sales_by_customers.sum()
```

```python
# Revisamos valores aleatorios del datasets
filas_aleatorias = sales_by_customers.sample(n=11)
print(filas_aleatorias)
```

```python
#Eliminamos la columna de 'subrubro' y la columna 'precio'.
sales_by_customers = sales_by_customers.drop(['subrubro', 'precio', 'id'], axis=1)
sales_by_customers
```

```python
# Eliminamos 'nomsub'
sales_by_customers = sales_by_customers.drop(['nomsub'], axis=1)
sales_by_customers
```

```python
# Separamos la tabla en dos tablas distinas, basandonos en el codigo, nomrub, nombre 
articles = sales_by_customers.loc[:,['rubro', 'nomrub', 'codigo', 'nombre']]
sales_articles = sales_by_customers.loc[:,['codigo', 'unidades','venta']]
```

```python
articles
```

```python
sales_articles
```

```python
#Guardamos ambas tablas en un csv
articles.to_csv('articles.csv', index= False)
sales_articles.to_csv('sales_articles.csv', index=False)
```

```python
#Insertamos nuestra tabla donde esta arrojada la informacion de cada compra que los clientes realizaron
but_costumers = pd.read_csv('VENTAS_POR_CLIENTES_REAL.CSV', sep=',', encoding='latin-1')
but_costumers
```

```python
#valores_azar = but_costumers.sample(n=11)
#print(valores_azar)
```

```python
#Vamos a eliminar las columnas innecesarias 'precios', 'nomrub', 'subrubro', 'nomsub', 
but_costumers = but_costumers.drop(['nomrub', 'subrubro', 'nomsub','precio'], axis=1 )
but_costumers
```

```python
#valores_azar1 = but_costumers.sample(n=11)
#print(valores_azar1)
```

```python
#Eliminamos la columna 'id' y 'nombre'. Además reemplazamos la columna 'rubro' por 'id_cliente';
# Y reemplazamos la columna 'codigo' por 'id_producto'
but_costumers = but_costumers.drop(['id', 'nombre'], axis=1)
```

```python
#Además reemplazamos la columna 'rubro' por 'id_cliente';
# Y reemplazamos la columna 'codigo' por 'id_producto'

but_costumers.rename(columns={'rubro':'ID_cliente'}, inplace=True)
but_costumers.rename(columns={'codigo':'ID_producto'}, inplace=True)
```
